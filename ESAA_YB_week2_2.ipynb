# 시본(Seaborn)
## 시각화를 위한 차트/그래프 유형
일반적으로 시각화는 통계 분석을 위한 시각화와 비지니스 분석을 위한 시각화로 나눠 볼 수 있다. 대부분의 시각화 차트는 두 영역 모두에서 사용되며, 비지니스 분석을 위한 시각화는 청중의 유형과 수준 그리고 설득이 필요한 다양한 환경에 따라서 적절하게 여러 가지 차트를 활용하도록 발전되어 왔다. 예를 들어, 특정 이산 값 칼럼에 따른 연속 값 결과를 나타낼 때는 바 차트, 라인 차트, 파이 차트, 도넛 차트 등을 이용할 수 있다. 특히 값의 비율을 좀 더 시각적으로 강조할 때는 파이 차트나 도넛 차트를 선택하는 게 더 나을 수 있다.

통계분석을 위한 시각화 또한 다양한 환경에 따라 적절한 차트를 활용한다. 하지만, 데이터 자체가 가지는 특성에 조금 더 집중하는 경향이 있다. 예를 들어, 특정 연속형 값이 정규 분포 형태가 아닌 왜곡된 분포를 갖고 있다거나 칼럼들 간의 상관관계, 이상치 값 도출 등 데이터 자체가 갖는 통계적인 특성을 설명하기 위해 시각화를 동원한다.

시본에는 많은 시각화 플롯을 지원하지만, 꼭 알아야 하는 시각화 플롯은 히스토그램, 바 플롯, 박스 플롯, 바이올린 플롯, 스캐터 플롯, 상관 히트맵이다. 이 플롯들은 데이터의 특성을 이해하는 데 기반이 되는 중요한 시각화 기법이다.

## 정보의 종유에 따른 시각화 차트 유형
1. 히스토그램
* 연속형 값에 대한 도수 분포를 나타낸다. X축 값은 도수 분포를 원하는 연속형 값의 구간, Y축 값은 해당 구간의 도수 분포(건수)를 나타낸다.
2. 바 플롯
* 특정 칼럼의 이산 값에 따른 다른 칼럼의 연속형 값(평균, 총합 등)을 막대 그래프 형태로 시각화한다. 수직 막대 그래프 적용 시, X축 값은 특정 칼럼의 이산 값, Y축 값은 다른 칼럼의 연속형 값으로 나타낸다.
3. 박스 플롯
* 연속형 값의 사분위 IQR와 최대, 최소, 이상치 값을 시각화한다. 보통 단일 칼럼의 연속형 값에 적용하지만, 이 연속형 값의 사분위를 다른 칼럼의 이산값 별로 시각화할 때 사용할 수 있다.
4. 스캐터 플롯
* 산점도로 불리며 2개의 연속형 값들을 X, Y 좌표상의 점으로 시각화하여 해당 값들이 어떻게 관계되어 있는지 나타낸다.
5. 상관 히트맵
* 다수의 연속형 칼럼들에 대해서 상호 간의 상관관계를 시각화한다.


여러 차트들 중에 바 플롯을 제외한 모든 플롯들이 기본적으로 연속형 칼럼 값에 대한 시각화를 지원한다. 시본은 X축과 Y축으로 구성된 이차원 축에서 데이터를 시각화해주므로 2개의 변수(칼럼)에 대한 정보를 기본적으로 표출해준다. 시본의 대부분 시각화 함수들은 hue라는 인자를 통해 또는 플롯의 유형에 따라 연속형 데이터의 정보를 다른 이산형 데이터 값으로 세분화된 정보로 표현할 수 있게 만들어주어 3개의 변수도 함께 정보로 시각화 할 수 있다. 이 외에도 박스 플롯, 바이올린 플롯, 스캐터 플롯 모두 hue 인자 또는 플롯 유형에 따라 이산형 값별로 연속형 값의 분포를 시각화할 수 있는 기능을 시본에서 제공한다.

## 히스토그램 (Histogram)
히스토그램은 데이터의 통계적인 분석을 위해 매우 많이 활용되며 특히 머신러닝 학습 데이터의 중요 피처들 값이 어떠한 분포를 갖고 있는지 확인하는 것은 모델 성능을 위해 중요하다.

히스토그램은 막대 차트처럼 보이지만, 연속형 값을 범위 또는 구간으로 그룹화해 개별 구간에 해당되는 데이터의 건수를 시각화해준다. 시본 히스토그램은 histplot( ) 또는 displot( ) 함수를 사용한다. 여기서 histplot( )은 Axes 레벨 함수이며, displot( )은 Figure 함수이다.

타이타닉 데이터 세트를 시본을 이용해 시각화 해보자.

```python
import pandas as pd

titanic_df = pd.read_csv('titanic_train.csv')
titanic_df.head(5)
```

시본의 histplot( ) 함수를 이용해 타이타닉 승객의 나이를 히스토그램으로 시각화해보자.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 시본에서도 plt.figure()의 figsize 인자를 입력하여 전체 Figure의 크기 조절
plt.figure(figsize=(8,3))

# DataFrame의 칼럼명을 자동으로 인식해서 xlabel 값을 할당. ylabel 값은 Count로 설정.
sns.histplot(titanic_df['Age'], bins=20)
plt.show()
```

20개의 bin 구간에 따라 연속으로 이어진 막대 그래프 형태의 도수 분포를 갖는 히스토그램이 만들어진다. 시본에서는 DataFrame을 입력값으로 받으면 자동으로 칼럼명을 해석해 X축명 혹은 Y축명에 할당한다. 그림에서 보듯이 별도의 xlabel( ) 호출 없이도 자동으로 X축명에 함수 인자로 들어온 Age 칼럼을 X축명에 할당하고 히스토그램 시각화이므로 Y축명에 도수를 나타내는 Count를 설정한다.

시본의 많은 시각화 함수들은 판다스 DataFrame과 통합되어 있어 X축, Y축 각각에 DataFrame의 칼럼명을 입력받아 처리할 수 있도록 설계되어 있다. 이를 위해 대부분의 시각화 함수들은 data, x, y를 각각 인자로 갖는데 data 인자는 시각화 대상 DataFrame, x 인자는 X축에 사용될 칼럼명, y 인자는 Y축에 사용될 칼럼명을 입력 받는다.

```python
sns.histplot(x='Age', data=titanic_df, bins=30, kde=True)
plt.show()
```

bins-30에 따라 30개의 구간별 막대 그래프 도수가 그려지고 kde=True에 따라 연속 확률분포 곡선까지 그려진다.

시본의 또 다른 함수인 displot( ) 함수는 Figure 레벨 함수이다. Figure 레벨 함수의 특징은 맷플롯립 API 사용을 최소화하고, 기본 맷플롯립에서 사용하는 기능들을 Figure 레벨 함수의 인자 등으로 대체하게 설계된다. 예를 들어, Figure 레벨 함수를 사용 시, Figure의 크기는 plt.figure( )로 조절할 수 없으며 해당 함수에서 인자로 Figure의 크기를 조절해야 한다.

displot( ) 함수를 이용해 Age 칼럼값의 히스토그램을 시각화해보자.

```python
import seaborn as sns

# seaborn의 figure레벨 그래프는 plt.figure( )로 Figure 크기를 조절할 수 있다.
plt.figure(figsize=(8,4))
sns.displot(x='Age', data=titanic_df, kde=True)
plt.show()
```

plt.figure( )를 이용해 가로의 길이를 8, 세로의 길이를 4로 설정하였지만, 오히려 세로가 더 긴 히스토그램이 만들어졌다. 이는 시본의 Figure 레벨 함수를 사용하면 plt.figure( )를 이용하여 Figure의 크기를 조절할 수 없기 때문이다.

displot( ) 함수의 Figure 크기를 조절하기 위한 인자로 height와 aspect가 있는데 height는 세로의 크기를 의미하며 aspect는 가로와 세로의 배율을 의미한다. 따라서, 가로의 길이는 height * aspect로 결정된다.

예를 들어, height가 4이고, aspect가 2인 경우, 가로 크기는 4*2=8로 결정된다.

```python
import seaborn as sns

sns.displot(titanic_df['Age'], kde=True, height=4, aspect=2)
plt.show()
```

가로 길이가 더 큰 Figure이 만들어진다.

시본의 Figure 레벨 함수는 그래프의 세부적인 변경이 다소 어렵다는 단점이 있지만, 여러 시각화 함수들을 한 번에 시각화 할 수 있다는 다양한 장점이 있다.

## 카운트 플롯
히스토그램이 연속형 값에 대해 구간에 따른 건수를 시각화한다면 카운트플롯은 이산형 값의 건수를 막대 그래프 형태로 시각화한다. 주로 카테고리성 칼럼 값별 건수를 시각화하는 데 사용된다. 카운트 플롯은 시본의 countplot( )을 사용한다.

```python
sns.countplot(x='Pclass', data=titanic_df)
plt.show()
```

타이타닉 데이터의 Pclass 값별 건수를 시각화하였다.
Pclass는 1, 2, 3 값을 가지고 있어, 타이타닉호의 1등실에는 200명이 족므 넘는 사람이, 2등실에는 200명이 안 되는 사람이, 3등실에는 거의 500명에 가까운 사람이 탑승했음을 알 수 있다.

## 바 플롯(barplot)
바 플롯(barplot)은 막대 그래프 형태의 플롯이다. 간단한 시각화이지만, 2차원 축 기반의 시각화에 널리 활용된다. 바 플롯은 X축 값이 이산형 값으로 값의 종류가 너무 많지 않을 때 유용하게 사용된다. 보통 Y축 값은 Y축에 해당하는 칼럼값의 평균이나 총합으로 표현되는 연속형 값을 가지며, 바 플롯을 수평 막대 그래프로 그리면 X축과 Y축이 바뀌게 되므로 Y축이 이산형 값, X축이 연속형 값을 갖도록 설정해야 한다.

시본을 이용해 타이타닉 데이터 세트의 Pclass 1, 2, 3 값별로 Age의 평균 값을 바 플롯을 이용해 표현해보자.

```python
# 자동으로 xlabel, ylabel을 barplot( )의 x 인자값이 pclass, y 인자값이 Age로 설정
sns.barplot(x='Pclass', y='Age', data=titanic_df)
plt.show()
```

맷플롯립과 다르게 pyplot 모듈의 xlabel( ). ylabel( ) 함수를 사용하지 않고도 X축명 Pclass, Y축명 Age를 자동으로 설정한다. barplot( ) 함수는 기본적으로 Y축 값의 평균 값을 나타낸다. 평균 외에도 총합, 중앙값 등을 나타낼 수도 있는데 이는 estimator 인자값을 설정하여 변경할 수 있다.

```python
sns.barplot(x='Pclass', y='Survived', data=titanic_df)
plt.show()
```

X축 값을 Pclass, Y축 값을 Survived로 설정했다. Survived 칼럼 값은 이산형 숫자값으로 0과 1, 두 개의 값(0은 사망, 1은 생존)으로 구성되어 있다. 하지만, 예제의 결과에서는 Y축 값이 0과 1 사이 소수점을 가진 실수로 표현되어 있다. 이는 바 플롯이 Y값의 평균, 즉 Survived의 평균을 표현했기 때문이다. 따라서 이 경우에는 선실 등급별 승객의 평균 생존율을 나타내게 된다.

barplot( ) 함수는 수직 또는 수평 막대 그래프를 시각화 선택을 orient 인자로 설정할 수 있다. orient가 v일 경우는 수직 막대 그래프, h일 경우는 수평 막대 그래프를 그린다. orient값을 별도로 지정하지 않는 경우, barplot( ) 함수는 입력된 X축 값과 Y축 값의 데이터 유형을 판단하여 자동으로 수직 또는 수평 막대 그래프를 그려주지만, X축과 Y값이 모두 숫자형 값이면 수직 막대 그래프를 우선하여 그린다.

바 플롯은 수직 막대 그래프의 경우 일반적으로 X축 값을 이산형 값으로 설정하며 이산형 값은 숫자값 또는 문자열값 모두 가능하다. 하지만, 수직 막대 그래프의 Y축 값을 문자열 값으로 설정해서는 안된다. 문자열 값의 평균, 총합은 정보로서 완전히 다른 의미이고, 시각적으로도 이해할 수 없기 때문이다. barplot( ) 함수를 사용할 때 만약 Y축 값을 문자열 값으로 입력하고 X축 값을 숫자형 값으로 입력하면 barplot( ) 함수는 자동으로 수평 막대 그래프로 시각화 한다.

```python
### y축을 문자값으로 설정하면 자동으로 수평 막대 그래프로 변환
sns.barplot(x='Pclass', y='Sex', data=titanic_df)
plt.show()
```

barplot( ) 함수가 x 인자로 숫자형 값인 Pclass, y인자로 문자열 값인 'Sex'를 입력받은 뒤 자동으로 이들 인자값의 데이터 유형을 파악하여 수평 막대 그래프로 전환해준다.

위 예제와 같이 X축, Y축 모두 이산값을 입력한다면 바 플롯으로 표현되는 정보를 시각적으로 잘못 이해될 수 있다. 또한, 바 플롯 생성 시 X축 값과 Y축 값을 모두 문자열 값으로 입력하면 바 플롯이 의미하는 정보의 표현 의도와 어긋나기에 오류를 발생시킨다.

```python
# x 인자로 문자열인 Name을, y 인자로 문자열인 Sex를 입력하므로 barplot은 오류를 발생
sns.barplot(x='Name', y='Sex', data=titanic_df)
```

수직 막대 그래프의 Y축 표현값을 평균이 아닌 총합으로 나타낼 수도 있다. 이때 사용되는 인자는 estimator이다. estimator=sum으로 설정하면 평균이 아닌 총합으로 표현할 수 있다.


```python
# estimator=sum을 적용해 평균이 아닌 총합으로 표현
sns.barplot(x='Pclass', y='Survived', data=titanic_df, estimator=sum)
plt.show()
```

Pclass 별로 Survived의 총합을 바 플롯으로 보여주므로 선실 등급별 생존자 수를 나타낸다.

### barplot( ) 함수의 hue 인자를 사용하여 시각화 정보를 추가적으로 세분화하기

일반적으로 2개의 축을 기반으로 한 2차원 평면 형태의 시각화는 X축과 Y축에 따른 정보를 시각화해주지만, 세부적인 추가 정보를 더해서 시각화할 수도 있다. 이러한 추가 정보는 좀 더 세분화된 비교 정보를 전달하는 데 매우 효과적이며, barplot( ) 함수는 hue 인자를 통해 이러한 세부 정보를 추가적으로 전달할 수 있다.

Pclass를 X축 값, Age를 Y축 값으로 설정하되, X축 Pclass별로 세부적인 추가 정보인 성별(Sex) 값에 따른 평균 나이를 시각화 해보자.

```python
# 아래는 Pclass가 X축값이며 hue 파라미터로 Sex를 설정
# 개별 Pclass 값별로 Sex에 따른 Age 평균 값을 구함
sns.barplot(x='Pclass', y='Age', hue='Sex', data = titanic_df)
plt.show()
```

Pclass가 1,2,3일 때 모두 2개의 막대 그래프로 나누어지는데 각각 Sex 값이 male일 때 파란색, female일 때 주황색 막대 그래프로 나누어졌다. 오른쪽 상단에는 범례로 Sex 칼럼의 개별 값들과 의미하는 색깔이 표시되어 있다. 기존의 Pclass 값별로만 비교하여 보여주던 정보에 개별 Pclass 값별로 Sex에 따른 세부적인 비교를 더하며 분석의 차원을 높이고 있다.

이번에는 Pclass 별 생존율을 더 세분화하여 Pclass에 따른 성별(Sex)별 생존율을 바 플롯으로 시각화해보자.

```python
# 개별 Pclass 값별로 Sex에 따른 Survived 평균값을 구함
# Pclass가 X축 값이며 Survived가 Y축 값. hue 파라미터로 Sex를 설정
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=titanic_df)
plt.show()
```

barplot( ) 함수만 아니라 시본의 여러 시각화 함수들은 hue 인자값을 이용해 추가적인 세부 정보를 시각화할 수 있도록 만들어주는 유용한 기능을 제공한다.

## 박스 플롯
박스 플롯은 상자 형태로 생긴 몸통 부분과 위, 아래로 길게 이어진 수염 혀애로 이뤄져 있다. 시본은 박스 플롯 시각화를 위해 boxplot( ) 함수를 제공한다. boxplot( ) 함수의 주요 인자로는 x, y, data가 있다. 박스 플롯은 분위수를 기반으로 하기에 연속형 값에 적용해야 의미 있는 정보가 될 수 있다. 박스 플롯은 연속형 값에 대한 IQR 분위와 최소/최대 그리고 이상치 정보를 시각화한다.

boxplot( ) 함수의 x 또는 y 인자로 연속형 값을 입력할 수 있는데, y에 입력 시 수직 박스 플롯을, x에 입력 시 수평 박스 플롯을 나타낸다.

```python
sns.boxplot(y='Age', data=titanic_df)
plt.show()
```

y 인자에 Age 칼럼을 설정해 Age 값의 분위수를 기반으로 한 수직 박스 플롯을 시각화 해주었다. 이처럼 기본적으로 박스 플롯은 단일 칼럼에 대한 분위수를 기반으로 하고 있지만, 추가적인 세분화 레벨에서도 적용이 가능하다.

예를 들어, 여러 개의 Pclass 값별로 Age에 대한 여러 개의 박스 플롯을 시각화 하기를 원한다면 y 인자에 Age를, 그리고 x 인자에 Pclass를 입력하면 된다. 추가적인 세분화 레벨로 여러 개의 박스 플롯을 표현하고자 한다면 x 인자는 이산형 값이 되어야 한다.

```python
sns.boxplot(x='Pclass', y='Age', data=titanic_df)
plt.show()
```

y 인자에 Age, x 인자에 Pclass를 입력하여 Pclass값 1,2,3 별로 Age에 대한 수직 박스 플롯을 그려주었다.

여기에 hue를 적용해 한 단계 더 추가적인 정보를 출력할 수 있다. hue 인자값으로 Sex 칼럼명을 입력하게 되면 Pclass값별로 세부적인 Sex 값인 male과 female에 따른 Age값의 박스 플롯들을 시각화할 수 있다.

```python
sns.boxplot(x='Pclass', y='Age', hue='Sex', data=titanic_df)
plt.show()
```

## 바이올린 플롯
바이올린 플롯은 히스토그램의 연속 확률 분포 곡선과 박스 플롯을 함께 시각화할 수 있다. 바이올린 플롯 역시 연속형 값에 적용해야 의미 있는 정보로 시각화될 수 있다. 시본은 바이올린 플롯을 위해 violinplot( ) 함수를 제공한다. violinplot( ) 함수는 boxplot( ) 함수와 동일하게 x 또는 y 인자로 연속형 값을 입력할 수 있는데, y에 입력시 수직 바이올린 플롯을, x에 입력 시 수평 바이올린 플롯을 나타낸다.

```python
# Age 칼럼에 대한 수직 바이올린 플롯 시각화
sns.violinplot(y='Age', data=titanic_df)
plt.show()
```

violinplot( )은 boxplot( )과 유사하게 여러 이산값 별로 여러 개의 바이올린 플롯들을 그릴 수 있다. 이는 여러 이산값벼로 여러 개의 히스토그램 연속 확률 분포 곡선들을 그려줄 수 있는 장점을 갖고 있다.

시본의 violinplot( ) 함수를 이용해 Pclass 별로 Age의 바이올린 플롯을 그려보자.

```python
# x축 값인 Pclass의 값별로 y축 값인 Age의 바이올린 플롯을 그림
# Pclass 값별 Age 데이터 분포를 비교하여 볼 수 있음
sns.violinplot(x='Pclass', y='Age', data=titanic_df)
plt.show()
```

1등실 승객의 나이는 30대 초반부터 40대 후반까지 비교적 넓은 범위에서 완만하게 집중되어 있지만, 2등실과 3등실 승객의 나이는 20대 중반부터 30대 초반까지의 상대적으로 좁은 범위에 집중되어 있음을 알 수 있다.

violinplot( ) 함수 역시 hue를 이용하여 한단계 더 추가적인 세부 정보를 제공할 수 있다. violinplot( ) 함수를 이용해 개별 Pclass 내에서 Sex 값별로 Age의 바이올린 플롯을 시각화 해보자.

```python
# x축값인 개별 Pclass 내에서 Sex값별로 y축 값인 Age의 바이올린 플롯을 그림
sns.violinplot(x='Pclass', y='Age', hue='Sex', data=titanic_df)
plt.show()
```

1등급일 경우는 남성의 경우가 여성보다 상대적으로 높은 나이대의 승객이 탑승했음을 알 수 있다. 반면의 2등급의 경우는 남성과 여성의 나이 분포가 큰 차이가 없으며, 3등급의 경우는 남성이 좀 더 청장년층에 집중되어 있고, 여성은 전 연령층에 분포되어 있음을 알 수 있다.

시본의 hisplot( ) 함수는 hue 기능을 제공하지만 violinplot( ) 함수가 특정 이산값에 해당하는 연속형 값의 데이터 분포를 훨씬 더 효과적으로 시각화해주기 때문에 데이터 분석 시 활용도가 높다.

## subplots를 이용하여 시본의 다양한 그래프를 시각화
시본의 다양한 그래프를 여러 개의 subplots 상에서 시각화 해보자.

시본의 시각화 함수를 subplots로 할당된 개별 Axes 객체에 적용하는 방식은 맷플롯립과 약간 다르다. 시본의 모든 Axes레벨 시각화 함수는 ax라는 인자를 가지고 있다. 시각화 함수 호출 시 이 ax 인자에 개별 Axes 객체를 할당하면 된다.

예를 들어 fig, axs = plt.subplots(nrows=1, ncols=3) 라고 한다면
sns.countplot(x=칼럼명, data=DataFrame, ax=axs[0]) 과 같이 ax 인자값으로 첫번째 로우 위치의 첫번째 칼럼 위치에 해당하는 Axes 객체 변수를 입력하게 되면 해당 위치의 subplot에 카운트 플롯을 시각화 해준다.

주요 이산형 칼럼인 Survived, Pclass, Sex의 건수를 시각화 해보자. 3개 컬럼이므로 세 개의 Axes 객체를 갖는 subplots를 생성하고 개별 Axes 객체에 시본의 countplot 함수를 적용한다.

```python
cat_columns = ['Survived', 'Pclass', 'Sex']

# nrows는 1이고, ncols는 칼럼의 개수만큼인 subplots를 설정
fig, axs = plt.subplots(nrows=1, ncols=len(cat_columns), figsize=(14,4))

for index, column in enumerate(cat_columns):
  print('index: ', index)
  # seaborn의 Axes 레벨 function들은 ax 인자로 subplots의 어느 Axes에 위치할지 설정
  sns.countplot(x=column, data=titanic_df, ax=axs[index])
```

건수를 표현할 칼럼들을 리스트 형태로 [Survived, Pclass, Sex]와 같이 생성한 뒤

fig, axs = plt.subplots(nrows=1, ncols=len(cat_columns), figsize=(14,4))를 호출하여 3개의 Axes를 생성하여 axs 객체 변수에 할당한다. axs[0], axs[1], axs[2] 각각은 왼쪽부터 생성되는 Axes 객체를 할당받게 되므로 루프를 돌면서 개별 칼럼에 대한 카운트 플롯을 시각화하되, ax 인자를 차례로 axs[0], axs[1], axs[2]를 입력하여 3개의 subplots를 출력하게 된다.

이번에는 subplots를 이용하여 Pclass, Sex, Embarked 3개의 이산형 칼럼별로 타깃 칼럼인 Survived 값의 평균값, 즉 생존율을 바 플롯으로 시각화 해보자.

```python
cat_columns = ['Pclass', 'Sex', 'Embarked']

# nrows는 1이고, ncols는 칼럼의 개수 만큼은 subplots을 설정
fig, axs = plt.subplots(nrows=1, ncols=len(cat_columns), figsize=(14,4))

for index, column in enumerate(cat_columns):
  # Seaborn의 Axes 레벨 function들은 ax 인자로 subplots의 어느 Axes에 위치할지 설정
  sns.barplot(x=column, y='Survived', data=titanic_df, ax=axs[index])
```

Pclass가 3일 때, 그리고 Sex가 male일 때 생존율이 매우 낮음을 알 수 있다. 이처럼 여러 개의 서브플롯을 이용하면 여러 개 칼럼들에 대한 시각화 정보를 한꺼번에 확인하고 비교해볼 수 있다. 특히 머신러닝 데이터 세트의 개별 피처 칼럼들이 서로 다른 타깃값별로 어떻게 구성되고 데이터가 분포되는지 확인하는 작업은 머신러닝의 데이터 탐색 과정에서 매우 중요한 요소이다.

다음으로는 연속형 칼럼들에 대해 타깃 칼럼인 Survived값에 따른 데이터 분포를 확인해보자. 두 개의 연속형 칼럼 Age와 Fare에 대해 Survived 값이 0일 때와 1일 때 각각 데이터 분포도를 확인하되, 개별 칼럼별로 두 개의 서브 플롯을 생성하여 왼쪽에는 Survived 값에 따른 바이올린 플롯을 오른쪽에는 Survived 값에 따른 히스토그램을 시각화 해보자.

```python
cont_columns = ['Age', 'Fare']

# 리스트로 할당된 칼럼들의 개수만큼 루프 수행
for column in cont_columns:
  # 왼쪽에는 바이올린 플롯, 오른쪽에는 히스토그램을 시각화. nrows는1, ncols=2인 서브플롯 생성
  fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
  # 왼쪽 Axes 객체에는 Survived 값 0, 1별 개별 칼럼의 바이올린 플롯 시각화
  sns.violinplot(x='Survived', y=column, data=titanic_df, ax=axs[0])
  # 오른쪽 Axes 객체에는 Survived 값 0, 1별 개별 칼럼의 히스토그램 시각화
  sns.histplot(x=column, data=titanic_df, kde=True, hue='Survived', ax=axs[1])
```

위 코드에서는 각 개별 칼럼별로 반복해서 코드를 호출하여 Figure와 서브플롯 Axes를 새롭게 생성한다. 매우 많은 컬럼들을 여러 가지 그래프를 가진 서브 플롯들을 많이 사용하여 시각화할 때는 모든 칼럼들의 서브플롯을 한꺼번에 만드는 방식보다는 위와 같이 서로 다른 그래프로 시각화할 서브플롯 두 개를 개별 칼럼별로 할당하고 이들을 모든 칼럼에 순차적으로 반복하여 적용하는 방식이 좀 더 쉬운 시각화 코드를 만들어준다.

Survived 값별로 Age의 바이올린 플롯을 살펴보면 Survived가 0일 때, 즉 사망한 승객의 나이대가 20대 초중반에 상당히 집중되어 있다. Survived가 1일 때, 즉 생존한 승객의 나이대는 그보다는 약간 높은 20대 중후반에 집중되어 있지만, 10대 이하와 40대 이상에서도 어느 정도 데이터가 분포되어 있다. 오른쪽의 히스토그램을 보더라도 Survived가 0인 경우는 20대 초중반의 나이대에 높은 도수 분포가 자리 잡고 있다. 앞에서 분석한 Sex별 생존율을 함께 결합해 살펴보면 남성이면서 20대 초중반의 승객이 상대적으로 많이 사망했음을 알 수 있다.

Fare의 경우 선실 등급이 높을수록 요금이 높기 때문에 Pclass와 밀접한 관계를 갖는다. 왼쪽의 바이올린 플롯에서는 Survived가 0인 경우 낮은 Fare값이 집중되어 있다. 이는 앞에서 분석한 Pclass 별 생존율과 크게 다르지 않다. 오른쪽 히스토그램 역시 Survived가 0일 때 매우 낮은 Fare값에 매우 높은 도수 분포가 자리하고 있음을 알 수 있다.

위와 같은 방식을 적절히 사용하면 분석해야 할 데이터의 칼럼 수가 많을 경우에도 이들 칼럼들과 타깃값과의 데이터 분포도 및 영향도를 시각적으로 이해할 수 있는 효과적인 방법이 될 것이다.

## 산점도, 스캐터 플롯(Scatter plot)
산점도는 스캐터 플롯(Scatter Plot)이라고 불리며, 보통 좌표상에 점을 표시하여 변수 간의 관계를 나타낸다. 일반적으로 산점도에 사용되는 X축 값, Y축 값 모두 연속형 숫자값을 적용해야 의미 있는 시각화 정보를 얻을 수 있다. 시본에서 산점도는 scatterplot 함수를 이용하여 시각화할 수 있다.

```python
# X축값으로 Age를, Y축값으로 Fare를 설정
sns.scatterplot(x='Age', y='Fare', data=titanic_df)
plt.show()
```

scatterplot을 이용해 타이타닉 데이터 세트의 Age와 Fare 간의 관계를 산점도로 시각화해 보았다. 산점도로 보면 주로 100 이하의 낮은 요금이 전 연령대에 걸쳐서 분포되어 있는 것으로 보인다. scatterplot 함수는 다른 시각화 함수와 마찬가지로 hue 인자를 지원한다. 이번에는 Age와 Fare로 표현되는 점을 사망/생존으로 추가적인 분리를 적용해보자.

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=titanic_df)
plt.show()
```

scatterplot( ) 함수는 hue 외에도 한 단계 더 세분화된 추가 정보를 제공할 수 있는데 바로 style 인자이다. 이번엔 hue='Survived'와 함께 style='Sex'를 입력해 시각화 해보자.

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', style='Sex', data=titanic_df)
plt.show()
```

X축 Age와 Y축 Fare로 표현되는 점들이 세부적으로 사망/생존은 파란색, 주황색으로 구분되며, 남성, 여성 성별은 남성이 원 모양으로, 여성은 엑스자 모양으로 구분됨을 알 수 있다. 이처럼 시본의 scatterplot( ) 함수는 hue 외에도 style 인자를 이용해 좀 더 세분화된 구분을 가능하게 한다.

## 상관 히트맵(Correlation Heatmap)
상관 히트맵은 다수의 속성들 간의 상관계수를 히트맵 형태로 나타낸다. 히트맵은 수치값을 온도를 연상시키는 다양한 색상으로 표현하는 것을 특징으로 한다. 상관계수는 두 속성들 간의 선형적인 연관 관계를 수치화한 값이다. 두 속성의 값이 서로 상관 없으면 0이며, 같은 방향으로 완전히 동일하면 1, 반대 방향으로 완전히 동일하면 -1 값을 갖는다.

상관 히트맵은 직관적으로 칼럼들 간의 상관도를 이해할 수 있게 해주는 휼륭한 시각화 기법이다. 상관 히트맵은 시본에서 heatmap( ) 함수를 통해 시각화할 수 있다. heatmap( ) 함수는 인자로 칼럼들 간의 상관계수를 갖는 DataFrame을 입력받아야 시각화가 가능하다. 판다스 DataFrame의 corr( ) 메서드를 호출하면 간단하게 상관계수를 갖는 DataFrame을 생성할 수 있다. 타이타닉 DataFrame에서 corr( ) 메서드를 호출하여 칼럼들 간의 상관계수를 갖는 DataFrame을 생성해보자.

```python
titanic_df.dtypes
```

```python
titanic_df.corr(numeric_only=True)
```

생성된 상관계수 DataFrame을 살펴보면 상단에 열 방향으로 칼럼들이 존재하며 왼쪽에 행 방향으로 역시 같은 칼럼들이 존재한다. 이는 칼럼과 칼럼 간의 상관계수값을 행과 열로 매핑하여 알 수 있게 해준다. 가령 행 방향에 위치한 Age 칼럼과 Survived 칼럼의 상관계수는 -0.077221이며 Age 칼럼과 Pclass 칼럼의 상관계수는 -0.369226이다. 가운데 대각선 방향으로 위치한 상관계수는 모두 1인데 이는 자기 자신과의 상관계수이기 때문이다.

이제 이 상관계수 DataFrame을 히트맵으로 표현해보자.

```python
# 상관계수를 DataFrame으로 구하고 이를 heatmap( )의 인자로 입력
corr = titanic_df.corr(numeric_only=True)
sns.heatmap(corr)
plt.show()
```

칼럼들 간의 상관계숫값에 해당하는 색상으로 상관 히트맵을 시각화했다. 맨 오른쪽에는 -1~1 사이의 숫자값에 해당하는 색깔 기준이 표현되어 있다. -1에 가까울 수록 검은색에 가까우며, 0으로 갈수록 빨간색이 되고, 1에 가까울 수록 밝은 베이지색으로 표현된다. Age 칼럼과 Pclass, Sibsp 칼럼은 검은색으로 음의 상관계수를 가짐을 알 수 있다. Survived 칼럼과 Pclass 칼럼은 검은색으로 음의 상관계수를 갖고 있다. 일반저긍로 상관계수는 연속형 숫작밧에서 해석되어야 한다. Survived와 Pclass는 숫자값이지만 카테고리성 칼럼이다. 하지만 Survived가 0일 때, Pclass가 1보다는 3인 값이 더 많고, Survived가 1일 때, Pclass가 1인 값이 더 많기 때문에 이러한 상관계수를 가지게 되었다.

heatmap( ) 함수는 여러 인자들로 보다 다양한 시각화를 할 수 있다. 먼저 cmap 인자는 color map으로서 히트맵의 색상을 변경할 수 있게 해준다. cmap 인자값을 설정하지 않으면 기본으로 'rocket'으로 설정된다.

```python
sns.heatmap(corr, annot=True, fmt='.1f', cbar=True)
plt.show()
```

상관계수 숫자값을 표시하면 칼럼 간의 상관계수를 좀 더 명확하게 확인할 수 있다. 결정값인 Survived는 Pclass와 -0.3, Fare와는 0.3의 상관계수를 갖는다. 또한, Pclass는 Age와 -0.4, Fare와 -0.5의 상관계수를 갖는다. 또한, Sibsp와 Parch는 0.4로 서로 밀접한 관계가 있음을 알 수 있다.

상관 히트맵은 칼럼 간의 상관관계를 이해할 수 있게 해주는 유용한 시각화 기능이다. 특히 칼럼들이 매우 많을 경우에 활용도가 매우 높으며 상관 히트맵을 통해 어떤 피처들이 타깃값과 밀접한 상관관계를 갖는지를 파악하여 해당 피처에 대한 피처 엔지니어링을 집중적으로 수행한다든지, 피처들끼리 상관도가 매우 높다면 상호 간의 높은 종속관계를 의심하여 특징 피처들을 제거하는 용도로 사용될 수 있다.

시본의 다양한 시각화 함수를 익힌다면 머신러닝 모델의 EDA를 수행하기에 충분한 시각화 역량을 갖게 될 것이다.

## 정리
맷플롯립과 시본을 이용해 여러 유형의 차트들을 시각화해보았다. 맷플롯립의 경우 pyplot 모듈의 두 가지 중요 요소인 Figure와 Axes의 차이와 이 두 객체의 핵심적인 역할에 대해 알 수 있었다. Figure는 그림을 그리기 위해 그림판의 크기를 조절하는 역할을 수행하지만, 실질적으로 그림을 그리는 역할은 Axes에서 수행한다. 뿐만 아니라 X축, Y축, 타이틀, 범례 등의 속성을 설정하는 데도 이용되므로 Axes 개체의 활용법을 이해하는 것은 맷플롯립과 시본을 활용하는 데에 중요하다.

pyplot 모듈의 plt.subplots( ) 함수를 이용하면 여러 개의 서브플롯들을 생성할 수 있고 이들을 이용하면 여러 개의 그래프들을 한꺼번에 시각화할 수 있다. 이때 할당되는 Axes 객체를 튜플이나 배열 인덱스로 접근할 수 있었으며, 시각화는 개별 서브플롯별로 할당된 Axes 객체의 시각화 함수들을 호출하여 수행할 수 있다.

시본은 맷플롯립보다 쉽게 다양한 시각화 차트를 만들 수 있다. 특히 판다스 DataFrame과 유연하게 통합되어 별도의 추가 코딩 없이도 축명이나 범례 등을 자동으로 설정할 수 있다. 시각화 차트/그래프의 유형은 많기에 정보의 종류나 유형에 따라 올바른 시각화 차트를 선택하는 것이 중요하다. 이를 위해서는 히스토그램, 카운트 플롯, 바 플롯, 박스 플롯, 바이올린 플롯, 스캐터 플롯, 상관 히트맵과 같이 머신러닝을 위해 핵심적으로 사용될 수 있는 시각화 플롯을 시본을 통해 구현해보았다. 개별적인 시각화 플롯들의 특징과 hue 인자를 사용해 시각화 정보를 추가적으로 세분화하는 방법, 시본을 여러 서브플롯들에 적용하는 기법들을 적절히 활용한다면 다양한 시각화의 요구사항을 만족하는 결과를 얻을 수 있다.
