```python
# 데이터 로드
from sklearn.datasets import fetch_openml
mnist=fetch_openml('mnist_784', version=1)
```

```python
# train, test split
X_train = mnist['data'][:60000]
y_train = mnist['target'][:60000]
X_test = mnist['data'][60000:]
y_test = mnist['target'][60000:]
```

* 랜덤포레스트에 적용

```python
# 랜덤포레스트에 적용 (n_estimators=10, random_state=42)
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=10, random_state=42)
rf_clf.fit(X_train, y_train)
```

```python
# 정확도 결과
from sklearn.metrics import accuracy_score
y_pred = rf_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

# 1. PCA

* PCA로 변환한 데이터 세트에 랜덤 포레스트 적용

```python
# PCA 라이브러리 불러오기
from sklearn.decomposition import PCA
```

```python
# PCA 적용 95% : n_components=0.95
pca = PCA(n_components=0.95)
X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
```

```python
# PCA 이후 랜덤포레스트 훈련
rf_clf2 = RandomForestClassifier(n_estimators=10, random_state=42)
rf_clf2.fit(X_train_reduced, y_train)
```

```python
# 정확도 결과 (X_test도 PCA 적용해야합니다.)
y_pred2 = rf_clf2.predict(X_test_reduced)
accuracy_score(y_test, y_pred2)
```

# 2. LDA

- LDA로 변환한 데이터 세트에 랜덤 포레스트 적용

```python
# LDA 라이브러리 불러오기

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# LDA 적용 : n_components=5
lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(X_train, y_train)
X_train_lda = lda.transform(X_train)
X_test_lda = lda.transform(X_test)
```

```python
# LDA 이후 랜덤포레스트 훈련
rf_clf3 = RandomForestClassifier(n_estimators=10, random_state=42)
rf_clf3.fit(X_train_lda, y_train)
```

```python
# 정확도 결과 (X_test도 LDA 적용해야합니다.)
y_pred3 = rf_clf3.predict(X_test_lda)
accuracy_score(y_test, y_pred3)
```
